```python
import numpy as np
array = np.array([[1, 2, 3], [4, 5, 6]])
print(array.ndim)
print(array.shape)
print(array.dtype)
```

```python
z = np.array(3, dtype=np.uint8)
print(z)
```

```python
double_array = np.array([1, 2, 3], dtype='d')
print(double_array)
```

```python
z = z.astype(np.float64)
print(z)
```

```python
sum = np.sum(array)
print(sum)
```

```python
mean = np.mean(array)
print(mean)
```

```python
std = np.std(array)
print(std)
```

```python
tiempos_api = np.array([195, 202, 198, 205, 200, 190])

media = np.mean(tiempos_api)
desviacion = np.std(tiempos_api)

print(f"Promedio: {media:.2f}ms")
print(f"Desviación Estándar: {desviacion:.2f}ms")
```
