Calculos Matriciales

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

#Suma
sum = A + B
print(A)
print(B)
print("\nSuma de matrices:")
print(sum)
```

```python
#Substraction
res = B - A
print("Resta de matrices:")
print(res)
```

```python
#Multiplication
product = np.dot(A, B)
print("Multiplicacion entre matrices:")
print(product)
```

```python
#Determinante
det_A = np.linalg.det(A)
#Inversa
inv_A = np.linalg.inv(A)
print(A)
print(det_A)
print(inv_A)
```

```python
#Resolver Ax = b
#Sistema de ecuaciones lineales
b = np.array([9, 11])
x = np.linalg.solve(A, b)
print(x)
```
