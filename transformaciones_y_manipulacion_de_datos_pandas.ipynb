Manipulacion de Columnas

```python
import pandas as pd

df = pd.read_csv('data_sets/online_retail.csv')
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print(df.head())
```

```python
df['HighValue'] = df['TotalPrice'] > 16
print(df['HighValue'].head(10))
```

```python
print(df.info())
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.info())
```

```python
df['DiscountedPrice'] = df['UnitPrice'].apply(lambda x: x*0.9)
print(df.head(3))
```

```python
def categorize_price(price):
    if price > 50:
        return 'High'
    elif price >= 20:
        return 'Medium'
    else:
        return 'Low'
    
df['PriceCategory'] = df['UnitPrice'].apply(categorize_price)
print(df.head())
```
