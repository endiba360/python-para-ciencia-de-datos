Estructuras y funciones

Funciones de resumen y estadisticas que podemo sutilizar con un DataFrame. Obten un resumen estadistico, calcula la media y mediana, haz una suma y conteo de valores.

```python
import pandas as pd

path = 'data_sets/online_retail.csv'
retail_data = pd.read_csv(path)

# Obtener nombres de columnas
columns_names = retail_data.columns
print(columns_names)
```

```python
num_rows, num_columns = retail_data.shape
print("Numero de filas", num_rows)
print("Numero de columnas", num_columns)
```

```python
daily_sales = retail_data['Quantity']
print(daily_sales)
print(daily_sales[2])
```

```python
summary = retail_data.describe()
print(summary)
```

```python
mean_value = daily_sales.mean()
print("La media es", mean_value)
```

```python
median_value = daily_sales.median()
print("La mediana es", median_value)
```

```python
total_sum_value = daily_sales.sum()
print("La suma total es", total_sum_value)
```

```python
count_values = daily_sales.count()
print("Total de valores no nulos: ", count_values)
```
